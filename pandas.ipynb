GroupBy

```python
import pandas as pd
technologies   = ({
    'Courses':["Spark","PySpark","Hadoop","Python","Pandas","Hadoop","Spark","Python","NA"],
    'Fee' :[22000,25000,23000,24000,26000,25000,25000,22000,1500],
    'Duration':['30days','50days','55days','40days','60days','35days','30days','50days','40days'],
    'Discount':[1000,2300,1000,1200,2500,None,1400,1600,0]
          })
df = pd.DataFrame(technologies)
```

```python
df
```

```python
df[['Courses', 'Fee', 'Discount']].groupby('Courses').sum()
```

```python
df.groupby(['Courses','Duration']).sum()
```

```python
groupedDF =df.groupby(by=['Courses'],dropna=False).sum() 
```

```python
groupedDF.sort_values('Courses',ascending=False)
```

```python
df.groupby('Courses')['Fee'].aggregate(['min','max'])
```

```python
df.groupby('Courses').aggregate({'Duration':'count','Fee':['min','max']})
```

Lambda functions

```python
values= [['Rohan',455],['Elvish',250],['Deepak',495],
         ['Soni',400],['Radhika',350],['Vansh',450]] 
 
# creating a pandas dataframe
df = pd.DataFrame(values,columns=['Name','Total_Marks'])
```

```python
df = df.assign(Percentage = lambda x: (x['Total_Marks']/500*100))
```

```python
df
```

```python
values_list = [[15, 2.5, 100], [20, 4.5, 50], [25, 5.2, 80],
               [45, 5.8, 48], [40, 6.3, 70], [41, 6.4, 90],
               [51, 2.3, 111]]
```

```python
df = pd.DataFrame(values_list, columns=['Field_1', 'Field_2', 'Field_3'])
```

```python
df = df.assign(Product=lambda x: (x['Field_1'] * x['Field_2'] * x['Field_3']))
```

```python
df
```

```python
values_list = [[15, 2.5, 100], [20, 4.5, 50], [25, 5.2, 80],
               [45, 5.8, 48], [40, 6.3, 70], [41, 6.4, 90], 
               [51, 2.3, 111]]
```

```python
df = pd.DataFrame(values_list, columns=['Field_1', 'Field_2', 'Field_3'],
                  index=['a', 'b', 'c', 'd', 'e', 'f', 'g'])
df
```

```python
df = df.apply(lambda x: np.square(x) if x.name == 'd' else x, axis=1)
```

```python
df
```

```python
df = df.apply(lambda x: np.square(x) if x.name in ['a','e','g'] else x, axis=1)
```

```python
values_list = [[1.5, 2.5, 10.0], [2.0, 4.5, 5.0], [2.5, 5.2, 8.0],
               [4.5, 5.8, 4.8], [4.0, 6.3, 70], [4.1, 6.4, 9.0],
               [5.1, 2.3, 11.1]]
df = pd.DataFrame(values_list, columns=['Field_1', 'Field_2', 'Field_3'],
                  index=['a', 'b', 'c', 'd', 'e', 'f', 'g'])
df = df.apply(lambda x: np.square(x) if x.name in ['b', 'f'] else x, axis=1)
df = df.assign(Product=lambda x: (x['Field_1'] * x['Field_2'] * x['Field_3']))
df
```
